```python
from pathogen import Pathogen
from segment import Segment
from sequence import Sequence
from random import choice
```

# Notes

In this simulation, we do the following:

- Initialize the environment with 2 viruses, of 2 segments each.
- At each time step, we guarantee 1 cycle of replication to happen, with only 1 progeny being produced per replication.
- At the end, we plot the network.

```python
# Initialize sequences
sequence_PB2 = Sequence(length=1000) #this is representative of segment 1, PB2
sequence_H1 = Sequence(length=1000) #this is representative of segment 4, H1
sequence_H2 = Sequence(length=1000) #this is representative of segment 4, H2
```

```python
# Initialize segments
segment_PB2 = Segment(segment_number=1, substitution_rate=1E-2, sequence=sequence_PB2)
segment_H1 = Segment(segment_number=4, substitution_rate=1E-1, sequence=sequence_H1)
segment_H2 = Segment(segment_number=4, substitution_rate=1E-1, sequence=sequence_H2)
```

```python
# Add segments to seed viruses
virus_H1 = Pathogen(segments=[segment_PB2, segment_H1], creation_time=0, convenient_id=1)
virus_H2 = Pathogen(segments=[segment_PB2, segment_H2], creation_time=0, convenient_id=2)
```

```python
# Add seed viruses to a list
viruses = []
viruses.append(virus_H1)
viruses.append(virus_H2)
```

```python
# Define the number of timesteps desired
timesteps = 10
```

```python
# Run through the timesteps, replicating one virus at each step
for i in range(1, timesteps):
    virus_to_replicate = choice(viruses)
    new_convenient_id = max(virus.convenient_id for virus in viruses) + 1
    new_virus = virus_to_replicate.replicate(current_time=i, convenient_id=new_convenient_id)
    viruses.append(new_virus)
```

```python
len(viruses)
```

```python
import networkx as nx
```

```python
# Create a new Directed Graph
transmission_graph = nx.DiGraph()
```

```python
# Add the nodes and edges to the Directed Graph
for virus in viruses:
    transmission_graph.add_node(virus, id=1)
    
    if virus.parent != None:
        transmission_graph.add_edge(virus.parent, virus)
```

```python
# Create positions for a time-restricted x-axis layout, with random integers as the y-axis.
positions = dict()
for virus in viruses:
    positions[virus] = (virus.creation_time, randint(0,20))

nx.draw(transmission_graph, pos=positions, with_labels=True)
```

```python
# Draw a simple circular layout.
nx.draw_circular(transmission_graph, with_labels=True)
```
